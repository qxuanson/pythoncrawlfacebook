## Làm sạch và tiền xử lý dữ liệu

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_df = pd.read_csv('Data/tintucvtv24.csv')
raw_df.info()
```

Lọc các dữ liệu cần thiết

```python
clean_df = raw_df[['post_id', 'post_text', 'time', 'video', 
                   'images_lowquality_description', 'comments', 
                   'shares', 'reaction_count']]
clean_df.info()
```

Tổ chức lại dữ liệu

```python
clean_df = raw_df[['post_id', 'post_text', 'time', 'comments', 'shares', 'reaction_count']]
clean_df = clean_df.rename(columns={'reaction_count': 'reaction', 'post_id': 'id', 'post_text': 'text'})
clean_df['media']=0
raw_df[['video','images_lowquality_description']] = raw_df[['video','images_lowquality_description']].astype(str)
for i in range(0,1000):
    if len(raw_df['images_lowquality_description'][i] )>7:
        clean_df['media'][i] = 1
    elif len(raw_df['video'][i])>7:
        clean_df['media'][i] = 2
    else:
        clean_df['media'][i] = 3
clean_df.info()
```

```python
clean_df['time'] = pd.to_datetime(clean_df['time'])
clean_df[['text']] = clean_df[['text']].astype(str)
clean_df.info()
```

Lọc thời gian bài viết

```python
start = pd.to_datetime('19-10-2023')
end = pd.to_datetime('19-11-2023')
clean_df = clean_df[(clean_df['time'] >= start) & (clean_df['time'] <= end)]
clean_df.info()
```
